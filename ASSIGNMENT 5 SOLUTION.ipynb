```python
import random
numbers = []
counter = 1
while counter <= 100:
    num = random.randint(14,70)
    numbers.append(num)
    counter +=1
print(numbers)
```

```python
import math
sqrtnumber =list(map(lambda x: (math.sqrt(x)),numbers))
print(f'THE SQUAREROOT OF THE RANDOM NUMBERS ARE {sqrtnumber}')

lognum =list(map(lambda x: (math.log(x)),numbers))
print(F'THE LOG OF THE RANDOM NUMBERS ARE {lognum}')

evennum =list(filter(lambda x: (x%2==0),numbers))
print(f'THE EVEN NUMBERS IN THE RANDOM NUMBERS ARE {evennum}')

oddnum =list(filter(lambda x: (x%2!=0),numbers))
print(f'THE ODD NUMBERS IN THE RANDOM NUMBERS ARE {oddnum}')
```
